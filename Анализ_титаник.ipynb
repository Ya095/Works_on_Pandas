```python
mdf = pd.DataFrame(index=df.index) # создание нового датафрейма с индексами равными датафрейму df
mdf['PassengerId'] = df['PassengerId'] # создаем колонку PassengerId в новом датафрейме
mdf['evenId'] = mdf['PassengerId'].apply(lambda x: x % 2 == 0) # создаем новую колонку в mdf, где будет bool значения, высчит. от колонки PassengerId в df. Четные - True, нечет. - False
```

```python
mdf
```

```python
pd.merge(df, mdf, how='inner') # соединяем таблицы по ключу методом inner.join
```

Изменение значений в dataframe и series

```python
tmp_df = df.copy(deep=True)
tmp_df
```

```python
tmp_df['Pclass'] = 1 # теперь все значения этой колонки = 1
tmp_df
```

```python
tmp_df['isAdult'] = tmp_df['Age'] >= 18 # создание новой колонки с bool значением
tmp_df
```

```python
tmp_df['isAdult'].value_counts() # количество совершеннолетних и нет
```

```python
tmp_df['isAdultSurvived'] = tmp_df['isAdult']
tmp_df.loc[tmp_df['Survived'] == 0, 'isAdultSurvived'] = False
```
